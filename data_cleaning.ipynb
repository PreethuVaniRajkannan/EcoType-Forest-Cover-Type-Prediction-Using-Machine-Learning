```python
numerical_cols = ['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology',
       'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways',
       'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
       'Horizontal_Distance_To_Fire_Points' ]
```

```python
#Outliers detection
plt.figure(figsize=(15, 7))
plt.suptitle("Boxplots before Outliers removal")
for i in range(0, len(numerical_cols)):
    plt.subplot(1, 12, i+1)
    sns.boxplot(y=df_data[numerical_cols[i]],color='purple')
    plt.tight_layout()
```

Observations:
Elevation - Very slight left skew, moderate distribution 
Aspect - normally distributed, since it is a degree based column it can be accepted
Slope - Right skew minimal outliers
Horizontal_Distance_To_Hydrology - Heavy Right skew more outliers
Vertical_Distance_To_Hydrology - High both ends - slight moderate spread
Horizontal_Distance_To_Roadways - Very slight left skew 
Hillshade_9am -  left skew
Hillshade_Noon -  left skew
Hillshade_3pm -  left skew
Horizontal_Distance_To_Fire_Points - right skew

Categorically encoded columns are ignored


Vertical_Distance_To_Hydrology shows strong positive skew and requires transformation. Hillshade_9am shows strong negative skew but may represent natural bounded distribution. Categorical features like Wilderness_Area and Soil type show extreme outliers but should not be treated as continuous variables.
Outliers are handled with IQR method.

```python
#APPLYING IQR METHODS to remove outliers in columns listed below
for col in ['Slope', 'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology','Hillshade_Noon', 'Hillshade_9am']:
    Q1 = df_data[col].quantile(0.25)
    Q3 = df_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    df_data[col] = df_data[col].clip(lower, upper)
```

```python
#categorical type change
df_data['Wilderness_Area'] = df_data['Wilderness_Area'].astype('category')
df_data['Soil_Type'] = df_data['Soil_Type'].astype('category')
```

```python
#Outliers detection
plt.figure(figsize=(15, 7))
plt.suptitle("Boxplots After Outliers removal")
for i in range(0, len(numerical_cols)):
    plt.subplot(1, 12, i+1)
    sns.boxplot(y=df_data[numerical_cols[i]],color='brown')
    plt.tight_layout()
```

outliers are handled now, and plot seem to be balanced.
